```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

# Загружаем данные

```python
train_data_path = 'contest_train.csv'
test_data_path = 'contest_test.csv'
```

```python
df_train = pd.read_csv(train_data_path, sep=',', header=0)
df_test = pd.read_csv(test_data_path, sep=',', header=0)
```

Фича FEATURE_256 бинарная, значений 0 больше, чем 1. Заполним NAN-значения этой фичи в тесте нулями

```python
df_test['FEATURE_256'].fillna((0.0), inplace=True)
```

Фичи FEATURE_238 и FEATURE_239 непрерывные. Заполним NAN-значения этой фичи в тесте средним

```python
df_test['FEATURE_238'].fillna((df_test['FEATURE_238'].mean()), inplace=True)
df_test['FEATURE_239'].fillna((df_test['FEATURE_239'].mean()), inplace=True)
```

## Удаляем фичи с NAN
Попробуем оценить распределение данных

Выберем все фичи без NAN, которые мы будем ужимать с помощью PCA

```python
not_null_df_train = (df_train.isna().sum() / len(df_train))
not_null_df_train = not_null_df_train[not_null_df_train == 0]

features = list(not_null_df_train.keys())
features.remove('ID')
features.remove('TARGET')
```

## Разделяем данные на трейн и вал

```python
target_df_train = df_train['TARGET'].copy()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_train.loc[:, features], target_df_train, test_size=0.2, random_state=49)
```

# Построение модели

## RandomForestClassifier

```python
target_df_train = df_train['TARGET'].copy()
```

```python
class_weight = {
    0: (target_df_train.value_counts()[1] + target_df_train.value_counts()[2]) / len(target_df_train),
    1: (target_df_train.value_counts()[0] + target_df_train.value_counts()[2]) / len(target_df_train),
    2: (target_df_train.value_counts()[0] + target_df_train.value_counts()[1]) / len(target_df_train),
}
```

```python
forest = RandomForestClassifier(criterion='gini', max_depth=7, min_samples_split=70, random_state=49,
                                n_estimators=100, n_jobs=8, verbose=2,
                               class_weight=class_weight, max_features=len(X_train.keys()))
```

```python
forest.fit(X_train, y_train)
```

```python
sorted(forest.feature_importances_)
```

```python
predicted_random_forest = forest.predict(X_test)
```

```python
print(classification_report(y_test, predicted_random_forest))
print(confusion_matrix(y_test, predicted_random_forest))
```

## DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(splitter='best', random_state=49, class_weight=class_weight, max_depth=50)
```

```python
tree.fit(X_train, y_train)
```

```python
sorted(tree.feature_importances_)
```

```python
predicted_tree = tree.predict(X_test)
```

```python
print(classification_report(y_test, predicted_tree))
print(confusion_matrix(y_test, predicted_tree))
```

## xgboost

```python
import xgboost as xgb
```

```python
df_target_xgb = y_train.copy()
df_target_xgb[df_target_xgb == 0] = class_weight[0] * 10
df_target_xgb[df_target_xgb == 1] = class_weight[1]
df_target_xgb[df_target_xgb == 2] = class_weight[2]
```

```python
xgboost_model = xgb.XGBClassifier(n_estimators=100, verbosity=2, random_state=49, n_jobs=8, 
                                  num_class=3, objective='multi:softmax', weights=df_target_xgb)
```

```python
xgboost_model.fit(X_train, y_train)
```

```python
predicted_xgboost_model = xgboost_model.predict(X_test)
```

```python
print(classification_report(y_test, predicted_xgboost_model))
print(confusion_matrix(y_test, predicted_xgboost_model))
```

```python
print(xgboost_model.feature_importances_)
```
